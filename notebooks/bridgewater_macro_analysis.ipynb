# 🌍 Bridgewater Associates Systematic Macro & Theme Analysis

This notebook explores the systematic **Risk-Parity** allocations of **Bridgewater Associates** (SEC CIK: `0001350972`).

We score their portfolio holdings across broad macroeconomic themes including Inflation Hedging, Defensive Quality, and Growth to model macro outlook positioning.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Premium dark style plotting configurations
plt.rcParams['figure.facecolor'] = '#121214'
plt.rcParams['axes.facecolor'] = '#1a1a1e'
plt.rcParams['text.color'] = '#eaeaea'
plt.rcParams['axes.labelcolor'] = '#eaeaea'
plt.rcParams['xtick.color'] = '#b0b0b5'
plt.rcParams['ytick.color'] = '#b0b0b5'
print("Macro themes notebook initialized!")
```

## 1. Define holdings with Macro Factors
We load Bridgewater's holdings along with systematic factors (beta, cash flow strength) to map macro thematic alignment.

```python
bridgewater_holdings = [
    {"ticker": "GLD", "weight": 12.0, "theme": "Inflation Hedging"},
    {"ticker": "XOM", "weight": 10.0, "theme": "Inflation Hedging"},
    {"ticker": "JNJ", "weight": 8.0, "theme": "Defensive Quality"},
    {"ticker": "MSFT", "weight": 5.0, "theme": "Secular Growth"}
]

df = pd.DataFrame(bridgewater_holdings)
df
```

## 2. Macro Thematic Scoring
Aggregating allocations by systemic macro themes.

```python
theme_df = df.groupby("theme")["weight"].sum().sort_values(ascending=False)

fig, ax = plt.subplots(figsize=(8, 5))
theme_df.plot(kind='bar', color=['#f29900', '#12a558', '#1a73e8'], ax=ax)
ax.set_title("Bridgewater Associates Thematic Macro Allocation Score", fontsize=14, fontweight='bold', pad=15)
ax.set_xlabel("Systemic Macro Theme", fontsize=12)
ax.set_ylabel("Thematic Weight (%)", fontsize=12)
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

# 💻 Citadel LLC Multi-Strategy portfolio Changes Analysis

This notebook analyzes the dynamic long portfolio changes of **Citadel LLC** (SEC CIK: `0001423053`), focusing on their rapid accumulation of semiconductor technology infrastructure.

We evaluate how their allocations align with global foundry indicators and market capitalization tiers.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Set dark style plotting configurations
plt.rcParams['figure.facecolor'] = '#121214'
plt.rcParams['axes.facecolor'] = '#1a1a1e'
plt.rcParams['text.color'] = '#eaeaea'
plt.rcParams['axes.labelcolor'] = '#eaeaea'
plt.rcParams['xtick.color'] = '#b0b0b5'
plt.rcParams['ytick.color'] = '#b0b0b5'
print("Citadel analysis configurations loaded!")
```

## 1. Define holdings Payload
We load Citadel's core long semiconductor and enterprise software allocations.

```python
citadel_tech_holdings = {
    "ticker": ["MSFT", "NVDA", "TSM", "ASML", "CELH"],
    "value_usd_m": [1200.00, 950.00, 600.00, 450.00, 150.00],
    "market_cap_tier": ["Mega Cap", "Mega Cap", "Mega Cap", "Mega Cap", "Mid Cap"],
    "domicile": ["United States", "United States", "Taiwan", "Netherlands", "United States"]
}

df = pd.DataFrame(citadel_tech_holdings)
df["allocation_weight"] = (df["value_usd_m"] / df["value_usd_m"].sum()) * 100
df
```

## 2. Domicile & Geography Analysis
Isolating international vs. domestic exposure weights.

```python
geo_df = df.groupby("domicile")["value_usd_m"].sum()

fig, ax = plt.subplots(figsize=(7, 7))
geo_df.plot(kind='pie', autopct='%1.1f%%', colors=['#f29900', '#12a558', '#1a73e8'], ax=ax, textprops={'color': '#eaeaea'})
ax.set_title("Citadel Tech Portfolio Domicile Distribution", fontsize=14, fontweight='bold', pad=15)
ax.set_ylabel("")
plt.tight_layout()
plt.show()
```

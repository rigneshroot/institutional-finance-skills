# 🍎 Berkshire Hathaway Sector Rotation & Capital Flow Analysis

This notebook tracks the transaction deltas and capital reallocations of **Berkshire Hathaway Inc.** (SEC CIK: `0001067983`) between Q2 and Q3 2024.

We isolate their Apple (AAPL) position reductions and Chevron (CVX) additions to model strategic rotation patterns.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Premium dark style plotting configurations
plt.rcParams['figure.facecolor'] = '#121214'
plt.rcParams['axes.facecolor'] = '#1a1a1e'
plt.rcParams['text.color'] = '#eaeaea'
plt.rcParams['axes.labelcolor'] = '#eaeaea'
plt.rcParams['xtick.color'] = '#b0b0b5'
plt.rcParams['ytick.color'] = '#b0b0b5'
print("Plotting configurations initialized!")
```

## 1. Define Historical holdings Comparison
We define holdings weights comparing consecutive quarters to isolate the rotation delta.

```python
data = {
    "ticker": ["AAPL", "BAC", "AXP", "KO", "CVX"],
    "weight_Q2": [50.50, 22.10, 12.40, 10.20, 4.80],
    "weight_Q3": [40.43, 18.51, 16.77, 13.88, 10.41]
}

df = pd.DataFrame(data)
df["rotation_delta"] = df["weight_Q3"] - df["weight_Q2"]
df
```

## 2. Plotting Rotation Deltas
Visualizing the exact capital shifts.

```python
fig, ax = plt.subplots(figsize=(10, 5))
colors = ['#d93025' if x < 0 else '#12a558' for x in df["rotation_delta"]]
df.plot(kind='bar', x='ticker', y='rotation_delta', color=colors, legend=False, ax=ax)
ax.set_title("Berkshire Hathaway Portfolio Allocation Shift (Q2 to Q3 2024)", fontsize=14, fontweight='bold', pad=15)
ax.set_xlabel("Stock Ticker", fontsize=12)
ax.set_ylabel("Weight Shift Delta (%)", fontsize=12)
plt.axhline(0, color='#eaeaea', linestyle='--', linewidth=0.8)
plt.tight_layout()
plt.show()
```
